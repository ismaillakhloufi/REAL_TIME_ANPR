```python
from ultralytics import YOLO
import torch
import matplotlib.pyplot as plt
from google.colab import drive

# Monter Google Drive
drive.mount('/content/drive')

# Charger un modèle YOLOv8 pré-entraîné
model = YOLO('yolov8n.pt')

# Entraînement du modèle avec des hyperparamètres améliorés
results = model.train(
    data='/content/drive/MyDrive/OCR FINAL/DATA/data.yaml',
    epochs=20,             # Augmentation des époques
    imgsz=640,
    batch=16,
    name='moroccan_plates',
    optimizer='Adam',      # Meilleur optimiseur pour la convergence
    lr0=0.001,            # Learning rate initial
    lrf=0.01,             # Facteur de réduction du learning rate
    weight_decay=0.0005,  # Régularisation L2
    dropout=0.3,          # Ajout de dropout pour éviter l'overfitting
    patience=10,          # Early stopping après 10 époques sans amélioration
    augment=True,         # Data Augmentation activée
    mosaic=1.0            # Augmentation basée sur Mosaic
)

# Visualisation des métriques
def plot_results(results):
    metrics = ['train/box_loss', 'val/box_loss', 'metrics/precision', 'metrics/recall', 'metrics/mAP_50']
    fig, axes = plt.subplots(1, len(metrics), figsize=(20, 5))

    for i, metric in enumerate(metrics):
        axes[i].plot(results.epoch, results[metric], label=metric, marker='o')
        axes[i].set_title(metric)
        axes[i].legend()
        axes[i].grid()

    plt.show()

# Afficher les métriques d'entraînement
plot_results(results)
```

```python
import matplotlib.pyplot as plt

# Access the first element of the results list (assuming it contains the Results object)
results_obj = results[0]

# Print results_obj.results_dict.keys() to check for the correct keys
print(results_obj.results_dict.keys())

# Access the loss and validation metrics using results_obj.results_dict
# Replace 'train/box_loss' and 'val/box_loss' with actual keys if different
train_losses = results_obj.results_dict.get('train/box_loss', [])
val_losses = results_obj.results_dict.get('val/box_loss', [])
epochs = results_obj.results_dict.get('epoch', [])

# Check if the lists are empty
print("train_losses:", train_losses)
print("val_losses:", val_losses)
print("epochs:", epochs)

# Create the plot if lists are not empty
if train_losses and val_losses and epochs:
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_losses, label='Training Loss')
    plt.plot(epochs, val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("Training data is missing. Please ensure the training process completed successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données du fichier results.csv
results_path = "runs/detect/moroccan_plates2/results.csv"  # Modifier si besoin
df = pd.read_csv(results_path)

# Vérifier les colonnes disponibles
print("Colonnes disponibles :", df.columns)

# Calculer la perte totale (somme des différentes pertes)
df['train/total_loss'] = df['train/box_loss'] + df['train/cls_loss'] + df['train/dfl_loss']
df['val/total_loss'] = df['val/box_loss'] + df['val/cls_loss'] + df['val/dfl_loss']

# Tracer les courbes de loss
plt.figure(figsize=(12, 5))

# Courbe de la perte d'entraînement et de validation
plt.plot(df['epoch'], df['train/total_loss'], label='Train Loss', color='blue')
plt.plot(df['epoch'], df['val/total_loss'], label='Validation Loss', color='red')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Courbes de Perte (Loss) - Entraînement et Validation')
plt.legend()
plt.grid()
plt.show()

# Tracer les métriques de validation (mAP, précision, rappel)
plt.figure(figsize=(12, 5))
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@50', color='green')
plt.plot(df['epoch'], df['metrics/precision(B)'], label='Precision', color='purple')
plt.plot(df['epoch'], df['metrics/recall(B)'], label='Recall', color='orange')

plt.xlabel('Epochs')
plt.ylabel('Score')
plt.title('Évolution des métriques de validation')
plt.legend()
plt.grid()
plt.show()
```
